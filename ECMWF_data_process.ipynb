```python
import os

# Get list of all .nc files in the directory
ncs = [f for f in os.listdir('IMD') if f.endswith('.nc')]
```

```python
from netCDF4 import Dataset as nco,num2date
import pylab as pl
import matplotlib.pyplot as plt
import numpy as np
#nf=nco('IMD/merged_IMD_2010_2024.nc')
nf=nco('sari.nc')

lat=nf.variables['latitude'][:]
lon=nf.variables['longitude'][:]

time = nf.variables['time']

tp=nf.variables['tp'][:]
date = num2date(time[:],time.units,only_use_cftime_datetimes=True)
dates=[d for i,d in enumerate(date)][-5:]
```

```python
import geopandas as gpd
from pyscissor import scissor
import numpy as np
import pandas as pd
from datetime import timedelta
a=[1,2,3,5,6,7,8]
for s in a:
    shp=gpd.read_file(f"D:\Masters\study_area\hechms\Subbasin_{s}.shp")
    shp_gcs = shp.to_crs("EPSG:4326")

    np.bool = np.bool_
    pys = scissor(shp_gcs['geometry'][0],lat,lon)

    weight_grid = pys.get_masked_weight() #=> returns a masked array containing weights
    print(weight_grid.shape)
    # get weighted average
    avg=[np.average(tp[i,:,:],weights=weight_grid) for i in range(time.shape[0])]
    
    df=pd.DataFrame(data={'Date':date,'Precipitation':avg})
    df['Date']=df['Date']+timedelta(hours=6)
    df.to_excel(f'Subbasin_{s}_IMDprecip.xlsx',index=False)
    
```
